```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.expand_frame_repr', False)
pd.set_option('display.precision', 2)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)

pd.set_option('display.float_format', lambda x: '%.5f' % x)
```

```python
dogs = pd.read_csv('data/dogs_cleaned_ready.csv')
cats = pd.read_csv('data/cats_cleaned_ready.csv')
```

```python
dogs['species'] = 'dog'
cats['species'] = 'cat'
```

```python
df = dogs.append(cats, ignore_index = True)
df.info()
```

```python
df.drop(columns=['declawed', 'breed_unknown', 'breed_primary', 'breed_secondary', 'color_primary', 'color_secondary', 'color_tertiary'], inplace=True)
df.info()
```

```python
sns.barplot(x='age', y='duration_as_adoptable', data=df, hue='species')
```

```python
sns.barplot(x='hasimage', y='duration_as_adoptable', data=df, hue='species')
```

```python
sns.barplot(x='size', y='duration_as_adoptable', data=df, hue='species')
```

```python
sns.barplot(x='goodwith_children', y='duration_as_adoptable', data=df, hue='species')
```

```python
sns.barplot(x='goodwith_cats', y='duration_as_adoptable', data=df, hue='species')
```

```python
sns.barplot(x='goodwith_dogs', y='duration_as_adoptable', data=df, hue='species')
```

```python
sns.relplot(x='distance', y='duration_as_adoptable', data=df, hue='species', kind='scatter', alpha=.5)
```

```python
sns.relplot(x='population', y='duration_as_adoptable', data=df, hue='species', kind='scatter')
```
